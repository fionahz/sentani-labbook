```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pysentani
import scipy as sp
```

```python
survey = pd.read_excel('../data-clean/sentani-merged-cleaned-2015-06-10.xlsx',na_values=[''])
print('number of entries =', len(survey))
print('number of columns =', len(survey.columns))
```

```python
pysentani.similar_columns(survey,'app_now')
```

```python
def access_check(app_column):
    no_access_mask = survey['access_type'] == 'no_access'
    app_mask = ((pd.notnull(survey[app_column])) & (survey[app_column] != 0))

    survey['sim_key'] = survey['simserial'].str[-5:]
    return survey[no_access_mask & app_mask][['village_name',
                                                 app_column,
                                                 'access_type',
                                                 'demand_point',
                                                 'sim_key']]
```

```python
access_check('app_now/fridge')
```

```python
access_check('app_now/fan')
```

```python
access_check('app_now/TV')
```

```python
access_check('app_now/rice_cooker')
```
